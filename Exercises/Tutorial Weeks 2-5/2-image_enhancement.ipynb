Exercise 1.2: Add salt and pepper noise to a grayscale image and check the result of previous filters

```python
# Select and image

img = cv2.imread(os.path.join(dataDir, 'car.jpg'))

# Convert it to grayscale

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Calculate the number of pixels of an image

h, w = gray.shape
number_pixels = h * w


num_alter_pixels = int(number_pixels * 0.1)

# Randomly select pixels to alter

pixels = np.random.randint(number_pixels, size=num_alter_pixels)

# choose color to convert (black = 0 or white = 255)

color = np.random.randint(2, size=num_alter_pixels) * 255

# Get pixel coordinates
y = pixels // w  # row index
x = pixels % w   # column index

# Ensure coordinates are within image boundaries
x = np.clip(x, 0, w-1)
y = np.clip(y, 0, h-1)

# Set salt of the pixels to black

for i in range(num_alter_pixels):
    img[y[i]][x[i]] = color[i]

# Show image
    
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyWindow('image')
```

Exercise 1.3: Reproduce gaussian blur with custom convolution (using [ndimage.convolve()](https://docs.scipy.org/doc/scipy/reference/generated/scipy.ndimage.convolve.html))

```python
from scipy import ndimage

# Open noisy image

img = cv2.imread(os.path.join(dataDir, 'coins_02.jpg'))

# Convert it to grayscale

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Normalize the image

img = img / 255

# Gaussian kernel

kernel = np.array([[1, 2, 1],
                   [2, 4, 2],
                   [1, 2, 1]])

# Normalize the kernel

kernel = kernel / np.sum(kernel)

# Apply convolution

img_conv = ndimage.convolve(img, kernel)

# Gaussian blur

img_guassian = cv2.GaussianBlur(img, (3,3), 0)


# Show image

cv2.imshow('image', img)
cv2.imshow('Gaussian', img_guassian)
cv2.imshow('Convolution', img_conv)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

Exercise 1.4: Define custom convolution (using [ndimage.convolve()](https://docs.scipy.org/doc/scipy/reference/generated/scipy.ndimage.convolve.html))

```python
from scipy import ndimage

# Open noisy image

img = cv2.imread(os.path.join(dataDir, 'coins_02.jpg'))

# Convert it to grayscale

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Normalize the image

img = img / 255

# Custom kernel

kernel = np.array([[0, 0, 0],
                     [0, 1, 0],
                     [0, 0, 0]])

# Apply convolution

img_conv = ndimage.convolve(img, kernel)

# Show image

cv2.imshow('image', img)
cv2.imshow('Convolution', img_conv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Exercise 1.5: Add salt and pepper noise to a colored image and apply the previous filters 

```python
img = cv2.imread(os.path.join(dataDir, "face_01.jpg"))

height, width, channels = img.shape

noise_density = 0.05

num_salt_pepper = np.ceil(noise_density * img.size)
coords = [np.random.randint(0, i, int(num_salt_pepper)) for i in img.shape]

# Apply salt and pepper noise to each channel separately
for channel in range(channels):
    img[coords[0], coords[1], channel] = np.random.choice(
        [0, 255], size=(int(num_salt_pepper))
    )


cv2.imshow("image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
for i in range(2):
    cv2.waitKey(1)
```

Exercise 1.6: Apply custom filter to colored image with noise

```python
# Apply custom filter to the colored image with noise

kernel = np.array([[0, 0, 0],
                     [0, 1, 0],
                     [0, 0, 0]])

img_conv = np.zeros((height, width, channels), dtype=np.uint8)

for channel in range(channels):
    img_conv[:, :, channel] = ndimage.convolve(
        img[:, :, channel], kernel, mode="constant", cval=0.0
    )

cv2.imshow("image", img)
cv2.imshow("Convolution", img_conv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 2. Histogram Equalization

```python
# Load low contrast image
img = cv2.imread(os.path.join(dataDir, 'face_lowContrast_01.jpg'), 0) # Change this, according to your image's path

cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

[Histograms Equalization](https://docs.opencv.org/master/d6/dc7/group__imgproc__hist.html#ga7e54091f0c937d49bf84152a16f76d6e)

```python
# Increasing contrast with Histograms Equalization
img_with_he = cv2.equalizeHist(img)

cv2.imshow('histogram_equalization', img_with_he)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

[Contrast Limited Adaptive Histogram Equalization](https://docs.opencv.org/master/d6/dc7/group__imgproc__hist.html#gad689d2607b7b3889453804f414ab1018)

```python
# Increasing contrast with CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
img_with_CLAHE = clahe.apply(img)

cv2.imshow('clahe', img_with_CLAHE)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Exercise 2.1: Apply Histogram Equalization to a colored image

```python
# Load low contrast color image

img = cv2.imread(os.path.join(dataDir, 'lowContrast_05.jpg'))

# Increase contrast with Histograms Equalization

img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Equalize the histogram of the V channel

img_hsv[:,:,2] = cv2.equalizeHist(img_hsv[:,:,2])

# Convert the HSV image back to RGB format

img_output = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)

cv2.imshow('image', img)
cv2.imshow('histogram_equalization', img_output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Exercise 2.2: Apply CLAHE to a colored image

```python
# Load low contrast color image

img = cv2.imread(os.path.join(dataDir, 'lowContrast_05.jpg'))

# Increase contrast with CLAHE

img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Equalize the histogram of the V channel

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))

img_hsv[:,:,2] = clahe.apply(img_hsv[:,:,2])

# Convert the HSV image back to RGB format

img_output = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)

cv2.imshow('image', img)
cv2.imshow('clahe', img_output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
